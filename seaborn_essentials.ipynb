```python
import seaborn as sns
%matplotlib inline
```

# loading dataset 

```python
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
```

```python
sns.distplot(tips_dataset['total_bill'], bins = 30, kde=True)
```

```python
sns.relplot(x='total_bill', y="tip", data=tips_dataset)
```

```python
sns.relplot(x='total_bill', y="tip", data=tips_dataset, hue='sex')
```

```python
sns.pairplot(tips_dataset, hue='sex', palette='coolwarm')
```

```python
sns.countplot(x='sex', data=tips_dataset)
```

```python
import pandas as pd
tips_dataset.groupby('sex').count()
```

# Pandas for visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d = {'A': np.random.rand(5),
    'B': np.random.rand(5),
    'C': np.random.rand(5),
    'D': np.random.rand(5),}
df = pd.DataFrame(d)
df
```

```python
df[['A']].hist(bins=30)
```

```python
df.plot.area()
```

```python
for i in range(4):
    sum = df.loc[i].sum().round(2)
    print(f'{i} sum = {sum}')
```

```python
spam = [df.loc[i].sum().round(2) for i in range(4)]
print(spam)
```

```python
df.plot.bar()
```

```python
df.plot.bar(stacked = True, alpha = 0.8)
```

```python
df.plot.bar(stacked = True, alpha = 0.1)
```
